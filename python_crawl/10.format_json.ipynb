# json使用注意事项
+ json中的字符串都是双引号
+ 如果不是双引号
    + eval：能实现简单的字符串和python类型的转换
    + replace：把单引号换成双引号

```python
import json
# import requests
from ten_requests_skills import parse_url
from pprint import pprint  # pretty print 格式化打印

url = "https://pm.3.cn/prices/mgets?origin=3&ext=01000000&skuids=1399903,27269595515,42010078901,2372300,100006388460,41220088267,2502890,35509125811,11803445806,25786381619&source=fashionv4"
html_str = parse_url(url)
print(type(html_str))

# json类型  ====>  python类型
ret1 = json.loads(html_str)
pprint(ret1)
print(type(ret1))

```

```python
# 读取json文件
with open("./douban1.json", "r", encoding="utf-8") as f:
    ret1 = f.read()
#     print(ret1)
    # json.loads 能把json类型的字符串转化为python类型的字符串
    ret2 = json.loads(ret1)
    print(ret2)
```

```python
json_str = json.dumps(ret2, ensure_ascii=False, indent=4)
print(json_str)
with open("./douban.json", "w", encoding="utf-8") as f:
    f.write(json_str)
```

# dumps：将python类型转换成json类型字符串


```python
test_dict = {'bigberg': [7600, {1: [['iPhone', 6300], ['Bike', 800], ['shirt', 300]]}]}
# 得到一个python类型的字典
print(test_dict)

# 将python类型的字典转换成json类型的字符串
json_str = json.dumps(test_dict)
print(json_str)
print(type(json_str))
```

# loads: 将json类型字符串转换成python字典


```python
python_dic = json.loads(json_str)
print(python_dic)
print(type(python_dic))

```
