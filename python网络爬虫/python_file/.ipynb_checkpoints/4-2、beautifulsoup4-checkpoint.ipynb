### 4.2.3方法选择器

`find_all()`  `find()`

<br>


`find_all(name, attrs, recursive, text, **kwargs)`

```python
# 1.name 根据节点名查询元素
html = '''
    <div class="panel">
    <div class="panel_heading">
        <h4>Hello</h4>
    </div>
    <div class="panel_body">
        <ul class="list" id="list-1">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-smart" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''

from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
uls = soup.find_all(name="ul")  # 返回一个类
print(type(uls[0]))
print(uls[0])
```

```python
# 对得到的element类还可以进一步进行find_all
for ul in uls:
    print(ul.find_all(name='li'))
    print(type(ul.find_all(name='li')))  # 返回tag类型
    
    
print('*'*100)
for ul in uls:
    print(ul.find_all(name='li'))
    for li in ul.find_all(name='li'):
#         print(li)
        print(li.string)
```

```python
# 2.attrs 根据属性进行查询
# attrs必须是一个字典


html = '''
    <div class="panel">
    <div class="panel_heading">
        <h4>Hello</h4>
    </div>
    <div class="panel_body">
        <ul class="list" id="list-1">
            <li class="element">Foo</li>
            <li class="element2">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-smart" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''
soup = BeautifulSoup(html, 'lxml')
panel_body = soup.find_all(attrs={'class':'panel_body'}) # 也可以用 find_all(class_='panel_body')
print(panel_body)

print('*'*100)

li_element2 = soup.find_all(name='li', attrs={'class':'element2'})
print(str(li_element2))
```

```python
# text 匹配文本，可以传入字符串，也可以传入正则表达式

import re
html = '''
<div class="panel">
<div class="panel_body">
<a>hello, this is a link</a>
<a>hello, this is a link, too</a>
</div>
</div>
'''

soup = BeautifulSoup(html, 'lxml')
text = soup.find_all(text='hello, this is a link')
text2 = soup.find_all(text=re.compile('link'))
print(text)
print(text2)
```

```python

html = '''
    <div class="panel">
    <div class="panel_heading">
        <h4>Hello</h4>
    </div>
    <div class="panel_body">
        <ul class="list" id="list-1">
            <li class="element">Foo</li>
            <li class="element2">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-smart" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''
soup = BeautifulSoup(html, 'lxml')
li_Bar = soup.find_all(name='li', text=re.compile('Bar'))
print(li_Bar)
li_element2_Bar = soup.find_all(name='li', class_='element2', text=re.compile('Bar'))
print(li_element2_Bar)
```

### 4.2.3 css选择器

```python
html = '''
    <div class="panel">
    <div class="panel_heading">
        <h4>Hello</h4>
    </div>
    <div class="panel_body">
        <ul class="list" id="list-1">
            <li class="element">Foo</li>
            <li class="element2">Bar</li>
            <li class="element">Jay</li>
        </ul>
        <ul class="list list-smart" id="list-2">
            <li class="element">Foo</li>
            <li class="element">Bar</li>
        </ul>
    </div>
</div>
'''

from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'lxml')
print(soup.select('ul#list-1'))
print(soup.select('#list-1 .element2'))

print('*'*100)
# 嵌套选择
uls = soup.select('ul')
# print(ul)
for ul in uls:
    li = ul.select('li')
    print(li)
    
print('*'*100)  
# 获取属性
for ul in uls:
    print(ul['id'])
    
# 获取文本
lis = soup.select('li')
for li in lis:
    print('li text:', li.get_text())
```
