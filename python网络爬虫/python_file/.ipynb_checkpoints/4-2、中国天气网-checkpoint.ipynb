```python
import requests
import json
from bs4 import BeautifulSoup
import time


class ChinaWeather(object):
    def __init__(self):
        self.url_temp = "http://www.weather.com.cn{}"  
        self.headers = {
            "Referer": "http://www.weather.com.cn/textFC/gat.shtml",
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36",
            "Cookie": "HttpOnly; HttpOnly; userNewsPort0=1; f_city=%E5%B9%BF%E5%B7%9E%7C101280101%7C; UM_distinctid=16d478db3ff2b9-0812bbf8e2256e-5373e62-151800-16d478db40048c; CNZZDATA1262608253=1442294579-1568859809-https%253A%252F%252Fwww.google.com%252F%7C1568859809; defaultCty=101280101; defaultCtyName=%u5E7F%u5DDE; vjuids=-1a79f248d.16d478df5bc.0.036f0680e1a77; vjlast=1568863549.1568863549.30; Hm_lvt_080dabacb001ad3dc8b9b9049b36d43b=1568863532,1568863553; Wa_lvt_1=1568863532,1568863553; Hm_lvt_24b77a56c77ce27d009359b331cde12e=1568863590; Hm_lvt_36dcc28125c1b7e65fa2190352951396=1568863615; Wa_lvt_7=1568863615; Hm_lpvt_36dcc28125c1b7e65fa2190352951396=1568863695; Wa_lpvt_7=1568863695; Hm_lpvt_24b77a56c77ce27d009359b331cde12e=1568863833; zs=101280101%7C%7C%7Cyd-uv; Hm_lpvt_080dabacb001ad3dc8b9b9049b36d43b=1568867628; Wa_lpvt_1=1568867628"
        }
        
    
    def parse_url(self, url):
        """解析网页，返回html"""
        try:
            response = requests.get(url, headers=self.headers)
            if response.status_code == 200:
                print("get: ",url)
                return response.content.decode()
            return None
        except requests.RequestException:
            print('获取失败')
            return None
    
    def get_a_day(self, tables):
        a_day_info = []
        for table in tables:
            trs = table.find_all('tr')[2:]
            province = table.select('td.rowsPan a')[0].string
            date = table.select('tr td[colspan="3"]')[0].string[:-2]
#             print(date)
            for tr in trs:
                tds = tr.find_all('td')
                city_td = tds[-8]
#                 if index == 0:
#                     city_td = tds[1]
                city = list(city_td.stripped_strings)[0]
                min_temp_td = tds[-2]
                min_temp = list(min_temp_td.stripped_strings)[0]
                max_temp_td = tds[-5]
                max_temp = list(max_temp_td.stripped_strings)[0]
                city_info = {'provice':province, 'city':city, 'date':date,  'min_temp':min_temp, 'max_temp':max_temp}
                a_day_info.append(city_info)
        print('得到一天的数据')
        return a_day_info
    
    
    def extract_data(self, html_str):
        all_days_info = []
        soup = BeautifulSoup(html_str, 'lxml')
        conMidtabs = soup.find_all(name='div', class_='conMidtab')
        for conMidtab in conMidtabs:
            tables = conMidtab.find_all(name='table')
            a_day_info = self.get_a_day(tables)
            all_days_info.extend(a_day_info)
        print('提取成功')
        return all_days_info
            
    def save_data(self, weather_data):
        """保存数据"""
        with open('../crawl_file/4-2.China_weather.txt', 'a+', encoding='utf-8') as f:
            for a_day_data in weather_data:
                f.write(json.dumps(a_day_data, ensure_ascii=False))
                f.write(',')
        print("保存成功")
        
        
    def run(self):
        """实现主程序"""
        # 1.获取url
        url_list = ['/textFC/hb.shtml', '/textFC/db.shtml',
                    '/textFC/hd.shtml', '/textFC/hz.shtml',
                    '/textFC/hn.shtml', '/textFC/xb.shtml',
                    '/textFC/xn.shtml']
        for index in url_list:
            url = self.url_temp.format(index)
            # 2.发送请求，获取相应的数据
            html_str = self.parse_url(url)
            # 3.提取数据
            all_days_info = self.extract_data(html_str)
            # 4.保存数据
            self.save_data(all_days_info)
            time.sleep(2)
        return all_days_info
            
```

```python
if __name__ == "__main__":
    China_weather = ChinaWeather()
    weather_data = China_weather.run()
    
```

```python
# 进行排序

def get_min_temp(city_info):
    min_temp = city_info['min_temp']
    return min_temp
min_temp = get_min_temp(weather_data[0])
# print(min_temp)

weather_data.sort(key=get_min_temp)
print(weather_data)
```
