```python
# é¡¹ç›®æ¼”ä¹ 
# çˆ¬å–å¾®åšæŸåšä¸»å‰10çš„å†…å®¹

from urllib.parse import urlencode
import requests
from pyquery import PyQuery as pq
from pymongo import MongoClient
def mongo_weibo():
    client = MongoClient()
    db = client['weibo']
    collection = db['weibo']

    return collection


base_url = "https://m.weibo.cn/api/container/getIndex?"
headers = {
    "Referer": "https://m.weibo.cn/u/2830678474",
    "User-Agent":
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36",
    "X-Requested-With": "XMLHttpRequest"
}


def get_page(page):
    params = {
        'type': 'uid',
        'value': '2830678474',
        'containerid': '1076032830678474',
        'page': page
    }

    url = base_url + urlencode(params)
    try:
        response = requests.get(url, headers)
        if response.status_code == 200:
            return response.json()
    except requests.ConnectionError as e:
        print("Error is ", e.args)


def parse_page(json):
    if json:
        items = json.get('data').get('cards')
        for item in items:
            item = item.get('mblog')
            weibo = {}
            weibo['id'] = item.get('id')
            weibo['text'] = pq(item.get('text')).text()
            weibo['attitudes'] = item.get('attitudes_count')
            weibo['comments'] = item.get('comments_count')
            weibo['reposts'] = item.get('reposts_count')
            yield weibo

def save_to_mongo(result):
    collection.insert_one(result)
    print('ä¿å­˜æˆåŠŸ')
```

```python
if __name__ == "__main__":
    collection = mongo_weibo()
    for page in range(1, 11):
        json = get_page(page)
        results = parse_page(json)
        for result in results:
            print(result)
            save_to_mongo(result)
```

```python
import requests
from urllib.parse import urlencode
from pymongo import MongoClient
import time


class Toutiao(object):
    def __init__(self, name):
        self.name = name  # èŽ·å–è¦çˆ¬å–çš„æœç´¢å†…å®¹ç›¸åº”çš„å…³é”®å­—
        self.base_url = "https://www.toutiao.com/api/search/content/?"
        self.headers = {
            "cookie":
            "tt_webid=6736143139453601287; WEATHER_CITY=%E5%8C%97%E4%BA%AC; tt_webid=6736143139453601287; csrftoken=1173d1e59061616313ba56775c3f5118; _ga=GA1.2.73970871.1568380906; s_v_web_id=9d78d95a5883b8e8a9363aa0386c9e3a",
            "referer":
            "https://www.toutiao.com/search/?keyword=%E8%A1%97%E6%8B%8D",
            "user-agent":
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36",
            "x-requested-with": "XMLHttpRequest"
        }
        # åˆå§‹åŒ–mongoDBå®¢æˆ·ç«¯ï¼Œå¹¶æŒ‡å®šcollectionä¸º toutiao
        client = MongoClient(host='localhost', port=27017)
        db = client['TouTiao']
        self.collection = db['toutiao']

    def parse_url(self, url):
        """å‘é€è¯·æ±‚ï¼ŒèŽ·å–å“åº”"""
        print(url)
        try:
            response = requests.get(url, headers=self.headers)
            if response.status_code == 200:
                return response.json()
        except requests.ConnectionError as e:
            print("Error is ", e.args)

    def get_data(self, html_json):
        """æ ¹æ®ä¼ è¿›æ¥çš„jsonå­—å…¸ï¼Œæå–ç›¸åº”çš„æ•°æ®"""
        if html_json['count']:  # åˆ¤æ–­æ–°é—»çš„æ•°é‡
            items = html_json['data']  
            for item in items:
                img_dict = {}
                if item.get('has_image', None):  # åˆ¤æ–­æ•°æ®ä¸­æ˜¯å¦å«æœ‰å›¾ç‰‡
                    img_dict['title'] = item.get('title', None)  # èŽ·å–å›¾ç‰‡çš„title
                    img_dict['image_list'] = item.get('image_list')  # èŽ·å–æ‰€æœ‰å›¾ç‰‡çš„urlï¼Œè¿”å›žä¸€ä¸ªå­—å…¸
                    yield img_dict

    def run(self):
        """å®žçŽ°ä¸»ç¨‹åº"""
        has_more = 1  # å…ˆå‡è®¾ç½‘é¡µæœ‰å¤šä¸ªajaxè¯·æ±‚
        page = 0  # åˆå§‹åŒ–ç½‘é¡µçš„é¡µç 
        while has_more:
            # æž„é€ url
            self.params = {
                "aid": "24",
                "app_name": "web_search",
                "offset": page * 20,
                "format": "json",
                "keyword": self.name,
                "autoload": "true",
                "count": "20",
                "en_qc": "1",
                "cur_tab": "1",
                "from": "search_tab",
                "pd": "synthesis",
                "timestamp": int(time.time() * 1000)  # æ ¹æ®ajaxè¯·æ±‚ï¼Œæ¨¡æ‹Ÿç›¸åº”çš„æ—¶é—´æˆ³
            }
            page += 1
            url = self.base_url + urlencode(self.params)
            # å‘é€è¯·æ±‚èŽ·å–å“åº”
            html_json = self.parse_url(url)
            if html_json['count']:
                has_more = html_json['has_more']
            # æå–æ•°æ®
            img_dict_generator = self.get_data(html_json)
            # ä¿å­˜æ•°æ®
            for img in img_dict_generator:
                try:
                    self.collection.insert_one(img)
                    print('ä¿å­˜æˆåŠŸ')
                except:
                    print("ä¿å­˜è¿‡äº†")
            time.sleep(2)
```

```python
import re
from pymongo import MongoClient
import urllib
import os

def mongo_collection():
    """èŽ·å–mongodb collectionï¼štoutiao  """
    client = MongoClient(host='localhost', port=27017)
    db = client['TouTiao']
    collection = db['toutiao']
    return collection

def get_data_from_mongo(collection):
    result = collection.aggregate([{"$project":{"_id":0}}])  # èŽ·å–å›¾ç‰‡æ•°æ®
    for img in result:
        title = img.get('title')
        file_name = re.sub(r'[:ï¼šï¼Œã€‚,\.?ï¼Ÿï¼!\|"â€â€œ]', "", title)  # åŽ»æŽ‰ä¸è§„èŒƒå­—ç¬¦ï¼Œé˜²æ­¢åœ¨å†™å…¥æ–‡ä»¶æ—¶å› ä¸ºæ–‡ä»¶åè€Œå‡ºé”™
        image_list = img.get('image_list')
        for index, image in enumerate(image_list, start=1):
            src = image.get('url')
            file_path = "../crawl_file/images/toutiao/" + file_name + str(index)+".jpg"
            if not os.path.exists(file_path):  # æ ¹æ®æ–‡ä»¶ååˆ¤æ–­æ˜¯å¦æ–‡ä»¶å·²ç»ä¸‹è½½è¿‡äº†
                urllib.request.urlretrieve(src, file_path)
                print('ä¿å­˜æˆåŠŸ' + file_path)

        
if __name__ == "__main__":
    collection = mongo_collection()
    get_data_from_mongo(collection)
    
        
    
```
